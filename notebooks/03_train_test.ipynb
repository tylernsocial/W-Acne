```python
from pathlib import Path
import pandas as pd
from sklearn.model_selection import train_test_split

PROJECT_ROOT = Path("..")
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

df = pd.read_csv(PROCESSED_DIR / "labels.csv")
df.head()
```

```python
df["label"].value_counts()
```

```python
# going to need to train the model, then validate whcich will fine tune and check the model during training and then test 
from sklearn.model_selection import train_test_split # splits dataset into random separate parts 

train_df, temp_df = train_test_split(
    df,
    test_size=0.30, # 30% goes into temp and 70 goes train = take full dataset and split it into 70% training and 30% temporary
    stratify=df["label"], # essentially allows for balancing despite uneven class sizes
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df["label"],
    random_state=42
)

print("Train:", len(train_df))
print("Validation:", len(val_df))
print("Test:", len(test_df))
```

```python
print("Train distribution:")
print(train_df["label"].value_counts())
```

```python
print("train distribution")
print(train_df["label"].value_counts())
print("validation distribution")
print(val_df["label"].value_counts())
print("test distribution")
print(test_df["label"].value_counts())
```

```python
train_df.to_csv(PROCESSED_DIR / "train.csv", index=False)
val_df.to_csv(PROCESSED_DIR / "val.csv", index=False)
test_df.to_csv(PROCESSED_DIR / "test.csv", index=False)
```
